```python
import pandas as pd
from pyproj import Proj, transform
```

```python
temp = pd.read_csv('data/Gangnam/speed/Gangnam_speed_2022.csv', index_col='Unnamed: 0')
idx = temp.columns
df = pd.DataFrame(columns=['LINK_ID', 'geometry'])
df
```

```python
link = pd.read_excel('C:/users/skang/desktop/data/  /2022/   _20233 .xlsx', engine='openpyxl')
#c = dict.fromkeys(adf[''])
proj_1 = Proj(init='epsg:5181')
proj_2 = Proj(init='epsg:4326')
for i in range(len(idx)):
    tempdf = link[link['LINK_ID']==idx[i]]
    temp = 'LINESTRING ('
    for j in range(len(tempdf)):
        x_, y_ = transform(proj_1, proj_2, tempdf['GRS80TM_X'].iloc[j], tempdf['GRS80TM_Y'].iloc[j])
        temp += str(x_) + ' ' + str(y_) + ', '
    temp = temp[:-2]
    temp += ')'
    df = df.append({'LINK_ID': idx[i], 'geometry': temp}, ignore_index=True)
df.to_csv('data/Gangnam/Gangnam_edges.csv', index='LINK_ID')
```
