```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
spd = pd.read_csv('data/Gangnam/speed/Gangnam_speed_2022.csv', index_col='Unnamed: 0')
```

```python
for i in range(len(spd.columns)):
    for j in range(len(spd.index)):
        if np.isnan(spd.iloc[j,i]):
            cnt = 0
            prev = j
            next = j
            while np.isnan(spd.iloc[prev,i]):
                prev -= 1
                cnt += 1
            while np.isnan(spd.iloc[next,i]):
                next += 1
                cnt += 1
            spd.iloc[j,i] = spd.iloc[prev,i] - ((spd.iloc[prev,i] - spd.iloc[next,i]) / cnt)
spd.to_csv('data/Gangnam/speed/Gangnam_speed_2022_interpolation.csv', index=False)
```
