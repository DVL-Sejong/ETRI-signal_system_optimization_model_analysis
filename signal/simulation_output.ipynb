```python
def _pretty_print(current, parent=None, index=-1, depth=0):
    for i, node in enumerate(current):
        _pretty_print(node, current, i, depth + 1)
    if parent is not None:
        if index == 0:
            parent.text = '\n' + ('\t' * depth)
        else:
            parent[index - 1].tail = '\n' + ('\t' * depth)
        if index == len(parent) - 1:
            current.tail = '\n' + ('\t' * (depth - 1))
```

```python
for i in range(15):
    idlist = []
    tree = ET.parse('data/Gangnam/Simulation/edge' + str(i) + '.xml')
    root = tree.getroot()
    interval = root.find('interval')
    for edg in interval.iter('edge'):
        idlist.append(edg.get('id'))
        
    col = ['sampledSeconds', 'traveltime', 'overlapTraveltime', 'density', 'laneDensity', 'occupancy',
      'waitingTime', 'timeLoss', 'speed', 'speedRelative', 'entered', 'left']
    df = pd.DataFrame(columns = col, index = idlist)
    
    for edg in interval.iter('edge'):
        for co in df.columns:
            df.loc[edg.get('id'), co] = edg.get(co)
            
    df['sampledSeconds'] = df['sampledSeconds'].astype(float)
    df['traveltime'] = df['traveltime'].astype(float)
    df['overlapTraveltime'] = df['overlapTraveltime'].astype(float)
    df['density'] = df['density'].astype(float)
    df['laneDensity'] = df['laneDensity'].astype(float)
    df['occupancy'] = df['occupancy'].astype(float)
    df['waitingTime'] = df['waitingTime'].astype(float)
    df['timeLoss'] = df['timeLoss'].astype(float)
    df['speed'] = df['speed'].astype(float)
    df['speedRelative'] = df['speedRelative'].astype(float)
    
    df['speed'] = round(df['speed'] * 3.6, 2)
    df['Avg_Vehicle_on_the_lane'] = df['sampledSeconds'] / 3600
    df['Avg_volume'] = df['speed'] * df['density']
    df['Total_travelled_distance'] = df['speed'] * df['sampledSeconds']
    df['Avg_waitingtime'] = df['waitingTime'] / df['Avg_volume']
    df['Edge_length'] = df['sampledSeconds'] / 3600000 / df['density'] * 1000000
    
    df.to_csv('data/Gangnam/Simulation/output' + str(i) + '.csv')
```
